```python
import requests
from bs4 import BeautifulSoup
import string
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
page_number = 1
processor =str(input("enter processor: "))
brand = str(input("enter brand: "))
gen = str(input("enter generation: "))
max_amount = str(input("enter max amount: "))
link = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off'
final_link = "{}&p%5B%5D=facets.processor%255B%255D%3DCore%2B{}&p%5B%5D=facets.brand%255B%255D%3D{}&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D{}&p%5B%5D=facets.processor_generation%255B%255D%3D{}th%2BGen&page={}".format(link,processor,brand,max_amount,gen,page_number)

page = requests.get(final_link)
soup = BeautifulSoup(page.content, 'html.parser')

all_laps = soup.findAll(class_="_2kHMtA")

urls = []
for a in range(len(all_laps)):
    l = all_laps[a].find(class_='_1fQZEK').attrs.get('href')
    lap_link = link + l
    urls.append(lap_link)
    
names = []
for a in range(len(all_laps)):
    name = all_laps[a].find(class_= '_4rR01T')
    names.append(name.text)
    
data = []
for a in all_laps:
    processor = None
    ram = None
    os = None
    storage = None
    dispaly = None

    ratings = a.find('div', class_='_3LWZlK')
    num_ratings = a.find('span', class_='_2_R_DZ')
    price = a.find('div',class_='_30jeq3 _1_WHN1')
    org_price = a.find('div',class_='_3I9_wc _27UcVY')

    all_specs = a.find('ul',class_= '_1xgFaf')
    sp = all_specs.findAll(class_="rgWa7D")

    for i in range(len(sp)):
        if('Processor' in sp[i].text):
            processor = sp[i]

        elif('RAM' in sp[i].text):
            ram= sp[i]

        elif('Operating' in sp[i].text):
            os = sp[i]

        elif('SSD' in sp[i].text or 'HDD' in sp[i].text):
            storage = sp[i]

        elif('Display' in sp[i].text):
            display = sp[i]

    data.append({'ratings': ratings.text if ratings is not None else np.nan,
                'num_ratings':num_ratings.text if num_ratings is not None else np.nan,
                'processor': processor.text if processor is not None else np.nan,
                'ram': ram.text  if ram is not None else np.nan,
                'storage': storage.text if storage is not None else np.nan,
                'display': display.text,
                'price': price.text,
                'original_price': org_price.text if org_price is not None else np.nan})
    
```

```python
data1 = pd.DataFrame({'links':urls,
                         'Names':names})

data1
```

```python
data2 = pd.DataFrame.from_dict(data) 
data2
```

```python
df_lappy = pd.concat([data1, data2], axis=1)
```
